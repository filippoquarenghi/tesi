```python
def split(img):
    cropped_img1 = img[int(img.shape[0]/2)+10:int(img.shape[0]/2)+74, int(img.shape[1]/2):int(img.shape[1]/2)+64]
    cropped_img2 = img[20:84, 20:84]
    return cropped_img1, cropped_img2
```

```python
def read_images_in2(path):
    #prototipo x
    img = io.imread(path[0])
    img_br, img_ul = split(img)
    x = np.dstack((img_br,img_ul))
    #prototipo y
    if 'CT' in path[0]:
        label=[0,0]
    if 'MRI' in path[0]:
        label=[1,1]
    if 'PET' in path[0]:
        label=[2,2]
    #lista nomi immagini croppate
    a, b = path[0].split('.')
    l = [a+'_br.png', a+'_ul.png']
    
    for img in path[1:]:
        #y
        if 'CT' in img:
            label.extend([0]*2)
        if 'MRI' in img:
            label.extend([1]*2)
        if 'PET' in img:
            label.extend([2]*2)
        a, b = img.split('.')
        l.append(a+'_br.png')
        l.append(a+'_ul.png')
        #x
        img = io.imread(img)
        img_br, img_ul = split(img)
        img = np.dstack((img_br,img_ul))
        x = np.dstack((x,img))
    x = np.rollaxis(x,-1)
    label = np.concatenate((label,))
    return x, label, l
```
